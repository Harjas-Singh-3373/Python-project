```python
x=[1,2,3,4,5]
y=[10,15,25,30,50]
```

```python
plt.plot(x,y,color='blue', marker='*')
plt.title('Simple line plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

copying the data from the assignment and made the line graph
plt.plot(x,y,color='blue', marker='*')--- it is given for the formation of the line plot
plt.title('Simple line plot')--- is given for adding the title of the graph
plt.xlabel('x')--- it is given for adding label to the x axis
plt.ylabel('y')--- it is given for adding label to the y axis
plt.show()--- is given so as to display the graph

```python
students=['John','Jane','Alice','Bob']
marks=[75,85,60,90]
plt.bar(students, marks,color='green')
plt.title('Bar Graph')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

The data for students and marks are copied from the asignment.
plt.bar(students, marks,color='green')--- it is given for the formation of the bar graph
plt.title('Bar Graph')--- it is given to add title for the graph
plt.xlabel('Students')--- it is given for adding the x label
plt.ylabel('Marks')--- it is given for the adding the y label
plt.show()---it is given so as to display the graph

```python
regions = ['North America', 'Europe', 'Asia', 'Other']
revenue = [45, 25, 20, 10]
col = ['green', 'blue', 'red', 'yellow']

plt.pie(revenue, labels=regions, colors=col)
plt.title('Pie Chart')
plt.legend(loc='upper left')
plt.show()
```

the data for the regions and revenue is copied from the assignment and the list of color is prefred randomly 
(color dosen't represent anly physical significance).
plt.pie(revenue, labels=regions, colors=col)--- it is given for the formation of the pie chart
plt.title('Pie Chart')--- it is given so as add title to the pie chart
plt.legend(loc='upper left')--- it is given to tell which color represents which continent
plt.show()--- it is given so as to display the graph

```python
import numpy as np
data = np.random.randint(1, 101, size=1000)
plt.hist(data, bins=10)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

data = np.random.randint(1, 101, size=1000)--- 1000 random numbers are generated between 1 to 100 from the given syntax
