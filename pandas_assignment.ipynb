```python
data={
    'Name':['Alice','Bob','Charlie','Diana','Eve'],
    'Age':[24,27,22,32,29],
    'Department':['HR','Finance','IT','Marketing','HR'],
    'Salary':[45000,54000,50000,62000,47000]
}
df=pd.DataFrame(data)
print(df.head())

print(df['Age'].describe())

print(df['Salary'].describe())

avg_salary=df[df['Department']=='HR']['Salary'].mean()
print("Avg salary of employe in HR depatement: ", avg_salary )
```

Data set has been copied from the assignment and is conveted into tabels using pd.DataFrame()
The first five elements are converted using df.head()
Statistical summary is given using df('cols').describe() for age and salary
Average of salary of HR is calculated by df['cols']['target'].mean()

```python
Bonus=df['Salary']/10
df1=df.assign(Bonus=Bonus)
print(df1)
```

a new column of bonus is prepared which is 10% of their actual salary by firstly dividing the salary by 10 
and then creating a new column by assigning it through df.assign(bonus=bonus) where LHS bonus means a new 
column is created and RHS bonus is that the value under the bonus column is addeed

```python
df2=df[( df['Age']>=25) & (df['Age']<=35)]
print(df2)
```

A new dataframe is prepared where the age of employes are are between 25 and 35 by using & operator 

```python
print(df.groupby('Department')['Salary'].mean())
```

df.groupby('cols')[].mean() is used to find the average of salary of employee working under same department

```python
df3=df.sort_values(by='Salary',ascending=True)
print(df3)
```

A new dataset is prepared where the salary of empolyes are in ascending order by using sort_values function
